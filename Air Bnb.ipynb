```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("compressed_data.csv")
```

# Checking Missing Value 

```python
df.isnull().sum()
```

# handling missing value

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')
```

```python
df.info()
```

```python
df.fillna({'reviews per month':0,'last review':df['last review'].min()},inplace = True)
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['license','house_rules'],errors='ignore')
```

```python
df.head()
```

```python
df['price'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

# remove duplicates

```python
df.drop_duplicates(inplace=True)
```

# descriptive stats

```python
df.describe()
```

# What is the distribution of listing prices?

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'],bins=50,kde=True,color='red')
plt.title("Distribution of listing price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

# How are  different types of room distributed?


```python
df['room type']
```

```python
plt.figure(figsize = (8,5))
sns.countplot( x = "room type", data = df,color = 'hotpink')
plt.title("Types Of Room Distributed")
plt.show()
```

# How are listings distributed across different neighbourhoods?

```python
plt.figure(figsize=(12,8))
sns.countplot(y= 'neighbourhood group',data=df,color='lightgreen',order=df['neighbourhood group'].value_counts().index)
plt.title("Number of Listing by Neighbourhood Group")
plt.xlabel("Count")
plt.ylabel("Neighbourhood Group")
plt.show()
```

# What is the relationship between price and room type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x= 'room type',y= 'price',hue='room type',data = df ,palette='Set2')
plt.title("Room Vs Price")
plt.xlabel("Room Type")
plt.ylabel("Price($)")
plt.legend(title='Room Type')
plt.show()
```
